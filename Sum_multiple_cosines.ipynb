```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

We analyze beating and oscilattion behavior when multiple cosine with close frequency are considered

# Sum of two cosines

```python
energies = [8,9] # in meV
```

```python
sim_time = 10000 # in fs
t = np.arange(0.,sim_time)
hplanck = 0.65821196*1e3 # using planck reduced in meV*fs
```

```python
f_sum = 0.
for e in energies:
    omega = e/hplanck
    f_sum += np.cos(omega*t)    
```

```python
plt.plot(t,f_sum)
```

There is (fast) oscillating term tha oscillates as $\frac{\omega_1+\omega_2}{2}$ and a (slow) modulation that oscillates as $\frac{\omega_1-\omega_2}{2}$

```python
omega_mod = (energies[1]-energies[0])/(2*hplanck)
f_mod = 2.*np.cos(omega_mod*t)
```

```python
plt.plot(t,f_sum)
plt.plot(t,f_mod)
```

# Analysis of the time dependence of the spin polarization

We analyze the time dependence of the spin polarization. In particular, we are interested in clarifying the behaviour of the polarization when the pulse end.

We consider the effects of different k sampling.

We choose a field of 100 fs od width and set the computation time to 2 ps.

## k = Gamma - no commutator

In this case there is single k point with no spin split. The gap is 1.1418. 
Heavy holes and Ligth holes are degenerate.
The pump frequency is tuned to the gap.

The time dependent behavior of the conduction spin magnetization is

```python
data_dir = 'rt_gamma_no-com/FixSymm/dephase_0.002-freq_1.418/'
mgn_file = data_dir+'o-dephase_0.002-freq_1.418.spin_magnetization'
mgn_dep2em3 = C.dict_parser(mgn_file)
```

```python
data_dir = 'rt_gamma_no-com/FixSymm/dephase_0.0001-freq_1.418/'
mgn_file = data_dir+'o-dephase_0.0001-freq_1.418.spin_magnetization'
mgn_dep1em4 = C.dict_parser(mgn_file)
```

```python
plt.plot(mgn_dep2em3['T'],mgn_dep2em3['Mc_z'],label='dephase-0.002')
plt.plot(mgn_dep1em4['T'],mgn_dep1em4['Mc_z'],label='dephase-0.0001')
plt.legend()
```

The time dependence seems to be not related to the level of dephasing. It should be due to the field induced variation of the density. 

```python
data_dir = 'rt_gamma_no-com/FixSymm/dephase_0.0001-freq_1.418-int_100/'
mgn_file = data_dir+'o-dephase_0.0001-freq_1.418-int_100.spin_magnetization'
mgn_dep1em4_int100 = C.dict_parser(mgn_file)
```

```python
plt.plot(mgn_dep1em4_int100['T'],mgn_dep1em4_int100['Mc_z'],label='int-100')
plt.plot(mgn_dep1em4['T'],mgn_dep1em4['Mc_z'],label='dephase-0.0001')
plt.legend()
```

```python
plt.plot(mgn_dep2em3['T'],mgn_dep2em3['Mv_z'],label='dephase-0.002')
plt.plot(mgn_dep1em4['T'],mgn_dep1em4['Mv_z'],label='dephase-0.0001')
plt.legend()
```

```python
plt.plot(mgn_dep2em3['T'],mgn_dep2em3['Ms_z'],label='dephase-0.002')
plt.plot(mgn_dep1em4['T'],mgn_dep1em4['Ms_z'],label='dephase-0.0001')
plt.legend()
```

## k = complete random grid

We repeat the same calculation with the two level of dephasing but using the complete random grid of 200 point.
In this way we can see if the presence of many k points produce an effect of decoherence that damps the spin polarization also for low dephasing.

```python
data_dir = 'rt_k200random_no-com/FixSymm/dephase_0.002-freq_1.55/'
mgn_file = data_dir+'o-dephase_0.002-freq_1.55.spin_magnetization'
mgn_dep2em3 = C.dict_parser(mgn_file)

data_dir = 'rt_k200random_no-com/FixSymm/dephase_0.0001-freq_1.55/'
mgn_file = data_dir+'o-dephase_0.0001-freq_1.55.spin_magnetization'
mgn_dep1em4 = C.dict_parser(mgn_file)
```

```python
plt.plot(mgn_dep2em3['T'],mgn_dep2em3['Mc_z'],label='dephase-0.002')
plt.plot(mgn_dep1em4['T'],mgn_dep1em4['Mc_z'],label='dephase-0.0001')
plt.axhline(y=0,linestyle='--',c='red')
plt.legend()
print('asympototic value = '+str(mgn_dep2em3['Mc_z'][-1]))
```

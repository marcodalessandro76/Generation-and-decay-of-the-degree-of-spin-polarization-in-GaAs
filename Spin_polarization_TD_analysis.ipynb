# Analysis of the time dependence of the spin polarization

We analyze the time dependence of the spin polarization. In particular, we are interested in clarifying the behaviour of the polarization when the pulse end.

We consider the effects of different k sampling, pulse intensity and values of the dephasing.

We choose a field of 100 fs od width and set the computation time to 1000 fs.

## k = Gamma - no commutator

```python
data_dir = 'rt_gamma_no-com/FixSymm/NETime_1000-freq_1.55-pol_circular/'
mgn_file = data_dir+'o-NETime_1000-freq_1.55-pol_circular.spin_magnetization'
mgn_1000 = C.dict_parser(mgn_file)
```

```python
data_dir = 'rt_gamma_no-com/FixSymm/NETime_1000-freq_1.55-int_100-pol_circular/'
mgn_file = data_dir+'o-NETime_1000-freq_1.55-int_100-pol_circular.spin_magnetization'
mgn_100 = C.dict_parser(mgn_file)
```

```python
data_dir = 'rt_gamma_no-com/FixSymm/NETime_1000-freq_1.55-int_2000-pol_circular/'
mgn_file = data_dir+'o-NETime_1000-freq_1.55-int_2000-pol_circular.spin_magnetization'
mgn_2000 = C.dict_parser(mgn_file)
```

```python
data_dir = 'rt_gamma_no-com/FixSymm/NETime_1000-dephase_0.01-freq_1.55-pol_circular/'
mgn_file = data_dir+'o-NETime_1000-dephase_0.01-freq_1.55-pol_circular.spin_magnetization'
mgn_lowdephase = C.dict_parser(mgn_file)
```

```python
plt.plot(mgn_1000['T'],mgn_1000['Mc_z'],label='1000')
plt.plot(mgn_100['T'],mgn_100['Mc_z'],label='100')
plt.plot(mgn_2000['T'],mgn_2000['Mc_z'],label='2000')
plt.plot(mgn_lowdephase['T'],mgn_lowdephase['Mc_z'],label='lowdephase')
plt.legend()
plt.ylim(0.49,0.5)
```
